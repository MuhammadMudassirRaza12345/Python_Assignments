```python

import csv
from sqlite3 import Time
filename = "python_a_2022.csv"
# time
from datetime import date, datetime

x = datetime.now()
date_str=x.strftime("%d-%m-%Y")
# print(date_str)

# date
import time

t = time.localtime()
current_time = time.strftime("%I:%M %p", t)
# print(current_time)

present=[]
absent=[]
# t=Date
# Time

with open(filename, "r") as f:
        data = csv.reader(f,delimiter=",")
        print("File read successfully", filename)
#       print("List of Students")
#             sno=roll_Number       
        for roll_Number,row in enumerate(data):
            roll_number = row[0]
            name = row[1]
            fname = row[2]
            email = row[3]
            phone = row[4]
        #     print(roll_number,name,fname, email, phone) 
#             print(sno,")",roll_number,name,fname, email, phone)
#             get present or absent 
            
            if roll_Number == 0:
                Date="Date"
                Time="Time"
                print(roll_number,name,fname, email, phone)
                absent.append([roll_number,name,fname,email,phone,Date,Time])
                present.append([roll_number,name,fname,email,phone,Date,Time])
                

            elif roll_Number >= 0:
                today_date=date_str
                current_time=current_time
                print(roll_number,name,fname, email, phone) 
                present_absent=input("Present ? (P/A)").upper()
                if present_absent=="A":
                        #we 
                        absent.append([roll_number,name,fname,email,phone,today_date,current_time])
                else:
                # pass
                        present.append([roll_number,name,fname,email,phone,today_date,current_time])
            

print("The students which are absent",absent,"\n")
print("The students which are present",present,"\n")

with open("present.csv","w") as f:
    writer=csv.writer(f)
    writer.writerows(present)

with open("absent.csv","w") as f:
        writer=csv.writer(f)
        writer.writerows(absent)
 

```
